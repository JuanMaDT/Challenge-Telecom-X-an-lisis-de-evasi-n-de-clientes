#ðŸ”§ TransformaciÃ³n

```python
from pandas import json_normalize
clientes = json_normalize(clientes.to_dict(orient="records")) #Normaliza para desanidar columnas
clientes
```

```python
#Renombrar columnas quitando los .
clientes.columns = clientes.columns.str.replace('.','_')
```

```python
clientes.info()
#Se encuentra que los cargos totales se deben llevar a num.
```

```python
#Convierte a numerico
clientes['account_Charges_Total'] = pd.to_numeric(
    clientes['account_Charges_Total'],
    errors = 'coerce'
)
```

```python
#Limpia variable objeto "Churn"
clientes['Churn'] = clientes['Churn'].replace('',pd.NA)
clientes
```

```python
clientes.info()
clientes.isna().sum()
```

```python
clientes.head()
```

```python
clientes['Cuentas_Diarias'] = clientes['account_Charges_Monthly'] / 30
#Divide entre 30 cada dato de la columna de mes
```

```python
#Paso de variables SI/NO
binarias = [
    'customer_Partner',
    'customer_Dependents',
    'phone_PhoneService',
    'account_PaperlessBilling'
]
for col in binarias:
  clientes[col] = clientes[col].map({'Yes':1,'No':0})
#Mapea para Yes = 1 - No = 0

clientes ['Churn'] = clientes['Churn'].map({'Yes':1,'No':0})
#Mapea para Yes = 1 - No = 0
```

```python
#Traduccion de columnas
clientes = clientes.rename(columns={
    'customer_tenure': 'Antiguedad_Meses',
    'account_Charges_Monthly': 'Cargo_Mensual',
    'account_Charges_Total': 'Cargo_Total'
})
```

```python
pd.crosstab(clientes['customer_gender'],
            clientes['Churn'],
            normalize = 'index')
#Porcentaje de Churn por genero
```

```python
def churn_por_categoria(col):
    return pd.crosstab(
        clientes[col],
        clientes['Churn'],
        normalize='index'
    ).sort_values(1, ascending=False)
```

```python
churn_por_categoria('account_Contract')
churn_por_categoria('account_PaymentMethod')
churn_por_categoria('internet_InternetService')
```
